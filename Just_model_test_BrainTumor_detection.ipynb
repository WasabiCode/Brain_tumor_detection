```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
model = load_model("/Users/vaclavknapp/Desktop/Coding/Brain_tumor_classification/models/Brain_Tumor_Detection_Model.h5")
```

```python
img = cv2.imread("y703.jpg")
plt.imshow(img)
plt.show()
```

```python
resize = tf.image.resize(img, (256,256))
```

```python
yhat = model.predict(np.expand_dims(resize/255, 0))
```

```python
if yhat > 0.5:
    plt.imshow(img)
    plt.title("This brain has tumor")
    plt.show
else:
    plt.imshow(img)
    plt.title("This brain has no tumor")
    plt.show
```
